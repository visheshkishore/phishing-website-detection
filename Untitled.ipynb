### Preprocessing data
#### Converting url into vector form using Count Vectorizer and gather words using tokenizer

#### RegexpTokenizer
A tokenizer that splits a string using regular expression, which matches either the tokens or the separators between tokens.

```python
tokenizer = RegexpTokenizer(r'[A-Za-z]+')
```

```python
phishing_data.URL[0]
```

```python
tokenizer.tokenize(phishing_data.URL[0])
```

```python
#### Tokenizing all urls in the dataset
phishing_data['tokenized_text'] = phishing_data.URL.map(lambda t: tokenizer.tokenize(t))
```

```python
phishing_data.sample(10)
```

```python
### Using Porter stemmer of nltk to stem words in urls

stemmer = PorterStemmer()
phishing_data['stemmed_text'] = phishing_data['tokenized_text'].map(lambda l: [stemmer.stem(word) for word in l])
```

```python
phishing_data.sample(7)
```

```python
print('Getting joiningwords ...')
phishing_data['text_sent'] = phishing_data['stemmed_text'].map(lambda l: ' '.join(l))
phishing_data.sample(10)
```

```python
#sliceing classes
bad_sites = phishing_data[phishing_data.Label == 'bad']
good_sites = phishing_data[phishing_data.Label == 'good']
```

```python
good_sites.sample(5)
```

### Plotting wordcloud of the urls of two classes seperately

```python
def plot_wordcloud(text):
    
    comment_words = '' 
    stopwords = set(STOPWORDS) 

    for val in text['text_sent']: 

        val = str(val) 
        tokens = val.split() 

        for i in range(len(tokens)): 
            tokens[i] = tokens[i].lower() 

        comment_words += " ".join(tokens)+" "

    wordcloud = WordCloud(width = 1000, height = 1000, 
                    background_color ='black', 
                    stopwords = stopwords, 
                    min_font_size = 10).generate(comment_words) 

    # plot the WordCloud image                        
    plt.figure(figsize = (10, 10), facecolor = None) 
    plt.imshow(wordcloud) 
    plt.axis("off") 
    #plt.title("Most common words in legitimate urls", title)
    plt.tight_layout(pad = 0) 

    plt.show()
```

### Plotting wordcloud of Legitimate URLs

```python
# Plotting wordcloud of legitimate URLs
plot_wordcloud(good_sites)
```

### Plotting wordcloud of Malicious URLs

```python
plot_wordcloud(bad_sites)
```

#### Tf-Idf Vectorizer to transform corpora of text to vector of tfidf values 

```python
tf = TfidfVectorizer()
features = tf.fit_transform(phishing_data.text_sent) #transform all text which we tokenize and stemed
```

```python
features[:20].toarray()
```

#### Splitting data in training and testing


```python
X_train, X_test, Y_train, Y_test = train_test_split(features, phishing_data.Label, test_size = 0.3,)
```

### Logistic regression



```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()
logreg.fit(X_train, Y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report

print('Accuracy score of the Logistic Regression classifier with default hyperparameter values {0:.2f}%'.format(accuracy_score(Y_test, logreg.predict(X_test))*100.))
print('----Classification report of the Logistic Regression classifier with default hyperparameter value----')
print(classification_report(Y_test, logreg.predict(X_test), target_names=['Phishing Websites', 'Legitimate Websites']))
```
